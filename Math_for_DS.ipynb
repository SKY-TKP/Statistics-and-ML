

## Introduction to Machine Learning (ML)

Machine Learning คือศาสตร์ที่ว่าด้วยการทำให้คอมพิวเตอร์สามารถเรียนรู้จากข้อมูลได้โดยไม่ต้องอาศัยการโปรแกรมแบบตายตัว เป้าหมายคือการสร้างระบบที่สามารถปรับปรุงประสิทธิภาพได้เองโดยอาศัยประสบการณ์  ML มีบทบาทสำคัญในหลากหลายด้าน เช่น การแพทย์, การเงิน, การตลาด, และวิทยาการคอมพิวเตอร์

**ประเภทของ Machine Learning:**

* **Supervised Learning:** เรียนรู้จากข้อมูลที่มี Label (คำตอบ) โดย Algorithm จะสร้างฟังก์ชันที่แมป Input กับ Output  
    * **Classification:**  จำแนกข้อมูลออกเป็นกลุ่มๆ
        * **ตัวอย่าง:**  จำแนกอีเมลว่าเป็น Spam หรือไม่, จำแนกประเภทของดอกไม้จากรูปภาพ, จดจำใบหน้า
        * **Algorithms:**
            * Logistic Regression:  ใช้ในการจำแนกข้อมูลออกเป็น 2 กลุ่ม โดยใช้ฟังก์ชัน Sigmoid ในการแปลง Linear Combination ของ Features ให้อยู่ในช่วง 0 ถึง 1 ซึ่งแทนความน่าจะเป็นของแต่ละ Class \newline
            * Support Vector Machine (SVM):  สร้าง Hyperplane ที่แบ่งข้อมูลออกเป็น 2 กลุ่ม โดยให้ Margin ระหว่าง Hyperplane กับข้อมูลแต่ละกลุ่มมีค่ามากที่สุด \newline
            * Decision Tree:  สร้าง Tree-like Model ที่ใช้ในการตัดสินใจ โดยแต่ละ Node แทน Feature, แต่ละ Branch แทน Decision Rule, และแต่ละ Leaf แทน Class หรือ Value \newline
            * Random Forest:  รวม Decision Tree หลายๆ ต้นเข้าด้วยกัน เพื่อลด Overfitting และเพิ่มประสิทธิภาพ \newline
            * K-Nearest Neighbors (KNN):  จำแนกข้อมูลโดยดูจาก K Neighbors ที่ใกล้ที่สุด
    * **Regression:**  ทำนายค่า Output ที่เป็นตัวเลขต่อเนื่อง
        * **ตัวอย่าง:**  ทำนายราคาบ้าน, ทำนายอุณหภูมิ, ทำนายยอดขาย, ทำนายความสูง
        * **Algorithms:**
            * Linear Regression:  สร้าง Linear Model ที่แสดงความสัมพันธ์ระหว่าง Input (Features) กับ Output \newline
            * Polynomial Regression:  สร้าง Polynomial Model ที่แสดงความสัมพันธ์ระหว่าง Input กับ Output \newline
            * Support Vector Regression (SVR):  คล้ายกับ SVM แต่ใช้ในการทำนายค่า Output ที่เป็นตัวเลขต่อเนื่อง \newline
            * Decision Tree Regression:  คล้ายกับ Decision Tree Classification แต่ใช้ในการทำนายค่า Output ที่เป็นตัวเลขต่อเนื่อง \newline
            * Random Forest Regression:  คล้ายกับ Random Forest Classification แต่ใช้ในการทำนายค่า Output ที่เป็นตัวเลขต่อเนื่อง

* **Unsupervised Learning:** เรียนรู้จากข้อมูลที่ไม่มี Label โดย Algorithm จะค้นหารูปแบบหรือโครงสร้างในข้อมูล  
    * **Clustering:**  จัดกลุ่มข้อมูลที่มีลักษณะคล้ายกัน
        * **ตัวอย่าง:**  จัดกลุ่มลูกค้า, จัดกลุ่มยีน, จัดกลุ่มเอกสาร
        * **Algorithms:**
            * K-Means Clustering:  แบ่งข้อมูลออกเป็น K Clusters โดยให้ข้อมูลในแต่ละ Cluster มีระยะห่างจาก Centroid ของ Cluster นั้นน้อยที่สุด  
            * Hierarchical Clustering:  สร้าง Hierarchy ของ Clusters โดยเริ่มจากแต่ละข้อมูลเป็น Cluster เดียว แล้วค่อยๆ รวม Clusters ที่ใกล้กันเข้าด้วยกัน  
            * DBSCAN:  จัดกลุ่มข้อมูลโดยอาศัย Density ของข้อมูล
    * **Dimensionality Reduction:**  ลดจำนวนตัวแปรในข้อมูล
        * **ตัวอย่าง:**  Principal Component Analysis (PCA), t-SNE
        * **ประโยชน์:**  
            * ลดความซับซ้อนของ Model
            * ลดเวลาในการประมวลผล
            * Visualization
            * ลด Noise ในข้อมูล

* **Reinforcement Learning:**  เรียนรู้โดยการลองผิดลองถูกในสภาพแวดล้อมแบบโต้ตอบ (Interactive Environment)  Agent จะเรียนรู้ที่จะเลือก Action ที่ดีที่สุดเพื่อให้ได้ Reward สูงสุด  
    * **ตัวอย่าง:**  สอนหุ่นยนต์ให้เดิน, สอน AI ให้เล่นเกม, ระบบควบคุมอัตโนมัติ, ระบบแนะนำเส้นทาง
    * **องค์ประกอบ:**  
        * **Agent:**  ผู้กระทำ (เช่น หุ่นยนต์, AI)
        * **Environment:**  สภาพแวดล้อมที่ Agent โต้ตอบด้วย
        * **State:**  สถานะปัจจุบันของ Environment
        * **Action:**  สิ่งที่ Agent ทำ
        * **Reward:**  รางวัลหรือบทลงโทษที่ Agent ได้รับจากการกระทำ

**ขั้นตอนการสร้าง Machine Learning Model:**

1. **การเตรียมข้อมูล (Data Preparation):** ขั้นตอนที่สำคัญที่สุด  
    * **Data Collection:**  เก็บรวบรวมข้อมูลจากแหล่งต่างๆ
    * **Data Cleaning:**  จัดการกับ Missing Value, Outlier, ข้อมูลที่ไม่ถูกต้อง
    * **Data Transformation:**  แปลงข้อมูลให้อยู่ในรูปแบบที่เหมาะสมกับ Algorithm เช่น Feature Scaling, Encoding Categorical Variables
    * **Feature Engineering:**  สร้าง Feature ใหม่ๆ จาก Feature เดิม
    * **Data Splitting:**  แบ่งข้อมูลออกเป็น Training Set, Validation Set, และ Test Set

2. **การเลือก Model (Model Selection):**  เลือก Algorithm ที่เหมาะสมกับปัญหาและข้อมูล  
    * พิจารณาประเภทของปัญหา (Classification, Regression, Clustering)
    * พิจารณาขนาดของข้อมูล, จำนวน Feature, ความซับซ้อนของ Model

3. **การฝึกฝน Model (Model Training):**  ปรับ Parameter ของ Model ให้เหมาะสมกับข้อมูลโดยใช้ Training Set  
    * **Loss Function:**  ฟังก์ชันที่ใช้วัดความผิดพลาดของ Model
    * **Optimizer:**  Algorithm ที่ใช้ปรับ Parameter เพื่อลด Loss Function เช่น Gradient Descent, Adam

4. **การประเมิน Model (Model Evaluation):**  วัดประสิทธิภาพของ Model โดยใช้ Validation Set  
    * **Classification Metrics:**  Accuracy, Precision, Recall, F1-Score, AUC
    * **Regression Metrics:**  Mean Squared Error (MSE), Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), R-squared

5. **การปรับปรุง Model (Model Tuning):**  ปรับปรุง Model ให้มีประสิทธิภาพดียิ่งขึ้น  
    * **Hyperparameter Tuning:**  ปรับ Parameter ของ Algorithm เช่น Learning Rate, Number of Trees
    * **Ensemble Methods:**  รวม Model หลายๆ Model เข้าด้วยกัน เช่น Bagging, Boosting
    * **Regularization:**  ป้องกัน Overfitting เช่น L1, L2


# Matrices Algebra
**เมทริกซ์ (Matrix)** คือ ตารางสี่เหลี่ยมของตัวเลข มีบทบาทสำคัญใน Linear Algebra และ Machine Learning

* **สมาชิก (Element):**  แต่ละตัวเลขในเมทริกซ์  
    *  $ a_{ij} $ คือ สมาชิกในแถวที่ $i$ หลักที่ $j$

* **มิติ (Dimension):**  จำนวนแถวและหลัก  
    * เมทริกซ์ขนาด $m \times n$ มี $m$ แถว $n$ หลัก

* **เมทริกซ์พิเศษ:**
    * **เมทริกซ์จัตุรัส (Square Matrix):**  เมทริกซ์ที่มีจำนวนแถวเท่ากับจำนวนหลัก
    * **เมทริกซ์เอกลักษณ์ (Identity Matrix):**  เมทริกซ์จัตุรัสที่มีสมาชิกบน Diagonal เป็น 1 และสมาชิกอื่นๆ เป็น 0  เช่น  
    \begin{align*}
I_2 &= \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}
\end{align*}
    * **เมทริกซ์ศูนย์ (Zero Matrix):**  เมทริกซ์ที่มีสมาชิกทุกตัวเป็น 0  เช่น  \begin{align*}
O &= \begin{bmatrix} 0 & 0 \\ 0 & 0 \end{bmatrix}
\end{align*}

**การดำเนินการกับเมทริกซ์:**
* **การบวก ลบ:**  บวก ลบ สมาชิกในตำแหน่งเดียวกัน
    *  $A + B = [a_{ij} + b_{ij}]$ เช่น  
       $
       \begin{bmatrix}
       1 & 2 \\
       3 & 4
       \end{bmatrix}
       +
       \begin{bmatrix}
       5 & 6 \\
       7 & 8
       \end{bmatrix}
       =
       \begin{bmatrix}
       6 & 8 \\
       10 & 12
       \end{bmatrix}
       $

* **การคูณด้วย Scalar:**  คูณ สมาชิกทุกตัวด้วย Scalar
    * $kA = [ka_{ij}]$  เช่น  
      2 x \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} = \begin{bmatrix} 2 & 4 \\ 6 & 8 \end{bmatrix}$

      


* **การคูณเมทริกซ์:**  คูณแถวของเมทริกซ์แรกกับหลักของเมทริกซ์ที่สอง  
    * $AB = C$  โดยที่  $c_{ij} = \sum_{k=1}^{n} a_{ik}b_{kj}$
    * **เงื่อนไข:**  จำนวนหลักของเมทริกซ์แรกต้องเท่ากับจำนวนแถวของเมทริกซ์ที่สอง
    * **ตัวอย่าง:**
      $
      \begin{bmatrix}
      1 & 2 \\
      3 & 4
      \end{bmatrix}
      \begin{bmatrix}
      5 & 6 \\
      7 & 8
      \end{bmatrix}
      =
      \begin{bmatrix}
      19 & 22 \\
      43 & 50
      \end{bmatrix}
      
      

* **Transpose:**  สลับแถวเป็นหลัก และหลักเป็นแถว
    * $(A^T)_{ij} = a_{ji}$  เช่น
      $
      \begin{bmatrix}
      1 & 2 \\
      3 & 4
      \end{bmatrix}^T
      =
      \begin{bmatrix}
      1 & 3 \\
      2 & 4
      \end{bmatrix}
      $

* **Determinant:**  ค่าคงที่ที่คำนวณจากสมาชิกของเมทริกซ์จัตุรัส  
    * ใช้หา Inverse, แก้ระบบสมการเชิงเส้น
    * **ตัวอย่าง:**
      $ det \begin{bmatrix} a & b \\ c & d \end{bmatrix}  = ad - bc $

* **Inverse:**  เมทริกซ์ที่คูณกับเมทริกซ์ต้นแบบแล้วได้เมทริกซ์เอกลักษณ์  
    * $AA^{-1} = A^{-1}A = I$
    * **เงื่อนไข:**  เมทริกซ์ต้องเป็นเมทริกซ์จัตุรัส และ Determinant ต้องไม่เท่ากับ 0
    * **ตัวอย่าง:**
      $
      \begin{bmatrix}
      2 & 1 \\
      1 & 1
      \end{bmatrix}^{-1}
      =
      \begin{bmatrix}
      1 & -1 \\
      -1 & 2
      \end{bmatrix}
      $

## Matrices Operation

เพื่อ transpose เมทริกซ์ เราสามารถใช้ T attribute หรือฟังก์ชัน np.transpose() ใน numpy ก็ได้


```python
import numpy as np

matrix = np.array([[1, 2, 3], [4, 5, 6]])

transpose_matrix_t = matrix.T
transpose_matrix_fn = np.transpose(matrix)

print("Original Matrix:")
print(matrix)

# หา transpose ของเมทริกซ์ด้วย T attribute
print("\nTranspose using T attribute:")
print(transpose_matrix_t)

# หา transpose ของเมทริกซ์ด้วยฟังก์ชัน
print("\nTranspose using transpose() function:")
print(transpose_matrix_fn)
```

เพื่อคูณเมทริกซ์ด้วยค่าคงที่ เราใช้เครื่องหมาย *

```python
import numpy as np

matrix = np.array([[1, 2, 3], [3, 4, 5]])

constant = -1

# คูณเมทริกซ์ด้วยค่าคงที่
multiplied_matrix = constant * matrix

print("Original Matrix:")
print(matrix)

print("\nMultiplied Matrix:")
print(multiplied_matrix)
```

เพื่อบวกและลบสองเมทริกซ์เข้าด้วยกัน เราใช้เครื่องหมาย + และ - ตามลำดับ


```python
import numpy as np

matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# บวกเมทริกซ์
result_matrix = matrix1 + matrix2  # ขนาดต้องเท่ากันก่อน

print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

print("\nResult:")
print(result_matrix)
```

เมทริกซ์จะบวกหรือลบกันได้นั้นต้องมีมิติที่เท่ากัน

```python
import numpy as np

matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[1, 2, 3], [3, 4, 5]])

print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

if matrix1.shape == matrix2.shape:
  result_matrix = matrix1 + matrix2
  print("\nResult:")
  print(result_matrix)
else:
  print("\nYou can only add or subtract two matrices if their dimensions are compatible")
```

## Matrices Multiplication

เราใช้ฟังก์ชัน np.dot() หรือเครื่องหมาย @ ในการคูณเมทริกซ์

```python
import numpy as np

matrix1 = np.array([[1, 2, 3], [3, 2, 1]])
matrix2 = np.array([[0, 2], [1, -1], [0, 1]])

# คูณเมทริกซ์ด้วยฟังก์ชัน
result_matrix_fn = np.dot(np.dot(matrix1, matrix2), matrix1) # np.dot() มักใช้ในการ dot product ของ vector มากกว่า

# คูณเมทริกซ์ด้วย @
result_matrix_at = matrix1 @ matrix2 @ matrix1  # แบบนี้นิยมมากกว่า

print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

print("\nResult of matrix multiplication using function:")
print(result_matrix_fn)

print("\nResult of matrix multiplication using @:")
print(result_matrix_at)
```

การคูณกันของเมทริกซ์นั้นไม่มีสมบัติการสลับที่





```python
import numpy as np

matrixA = np.array([[1, 2, 3], [3, 2, 1]])
matrixB = np.array([[0, 2], [1, -1], [0, 1]])

# หาผลคูณ AB และ BA
result_matrix_AB = matrixA @ matrixB
result_matrix_BA = matrixB @ matrixA

print("Matrix A:")
print(matrixA)

print("\nMatrix B:")
print(matrixB)

print("\nResult of matrix multiplication A * B:")
print(result_matrix_AB)

print("\nResult of matrix multiplication B * A:")
print(result_matrix_BA)
```

เมทริกซ์จะคูณกันได้นั้น จำนวน column ของเมทริกซ์แรกต้องเท่ากับจำนวน row ของเมทริกซ์หลัง

```python
import numpy as np

matrix1 = np.array([[1, 2], [3, 2]])
matrix2 = np.array([[0, 2, 1], [1, -1, 3], [0, 1, 2]])

print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

#print(matrix1 @ matrix2)

if matrix1.shape[1] == matrix2.shape[0]:
  print("\nResult after subtraction:")
  print(matrix1 @ matrix2)
else:
  print("\nYou can only multiply two matrices if their dimensions are compatible.")
```

## Linear Transformation


scaling คือการย่อหรือขยายเวกเตอร์ด้วยการคูณค่าคงที่

## Matrix Equation

```python
import numpy as np
import numpy.linalg

matrix = np.array( [[ 0.8, 0, 0.3], [ 0.1, 0.85 ,0], [ 0, 0.1, 0.7]])

# หา inverse ของเมทริกซ์
matrix_inverse = np.linalg.inv(matrix)

print("Original Matrix:")
print(matrix)

print("\nInverse Matrix:")
print(matrix_inverse)

print()
print(matrix_inverse @ matrix)
```

เราสามารถแก้ matrix equation Ax = b เพื่อหา x ได้ด้วยฟังก์ชัน np.linalg.solve() หรือคูณด้วย inverse

```python
import numpy as np
import numpy.linalg

A = np.array( [[0.8, 0, 0.3], [0.1, 0.85 ,0], [0, 0.1, 0.7]])
b = np.array([40, 10, 50])

matrix_inverse = np.linalg.inv(A)

# หาคำตอบของสมการเมทริกซ์ด้วยการคูณด้วย inverse
solution_inverse = matrix_inverse @ b

# หาคำตอบของสมการเมทริกซ์ด้วยฟังก์ชัน
solution_function = np.linalg.solve(A,b)

print("Solution vector by slove function:")
print(solution_function)

print("\nSolution vector by inverse function:")
print(solution_inverse)
```

# Linear Algebra
**พีชคณิตเชิงเส้น (Linear Algebra)** คือ สาขาของคณิตศาสตร์ที่ศึกษาเกี่ยวกับเวกเตอร์, ปริภูมิเวกเตอร์,  การแปลงเชิงเส้น, และระบบสมการเชิงเส้น  เป็นพื้นฐานที่สำคัญของ Machine Learning

**เวกเตอร์ (Vector):**  ปริมาณที่มีทั้งขนาดและทิศทาง  
* สามารถเขียนแทนด้วยเมทริกซ์ 1 มิติ (Column Vector) หรือ เมทริกซ์ 1 แถว (Row Vector)  เช่น  $ \vec{v} = \begin{bmatrix} 2 \\ 3 \end{bmatrix} $

**ปริภูมิเวกเตอร์ (Vector Space):**  เซตของเวกเตอร์ที่มีคุณสมบัติพิเศษ  
* **คุณสมบัติ:**  ปิดภายใต้การบวกและการคูณด้วย Scalar  (ผลบวกของเวกเตอร์ใดๆ ในเซต และผลคูณของเวกเตอร์ใดๆ ในเซตกับ Scalar ใดๆ จะต้องอยู่ในเซตนั้นด้วย)

**การแปลงเชิงเส้น (Linear Transformation):**  ฟังก์ชันที่แปลงเวกเตอร์หนึ่งไปเป็นอีกเวกเตอร์หนึ่ง  
* **คุณสมบัติ:**  รักษาการบวกและการคูณด้วย Scalar  
    * $T( \vec{u} + \vec{v}) = T(\vec{u}) + T( \vec{v})$
    * $T(k \vec{u}) = kT( \vec{u})$
* **ตัวอย่าง:**  การหมุน, การสะท้อน, การยืดหด, การฉาย

**ระบบสมการเชิงเส้น (System of Linear Equations):**  ชุดของสมการเชิงเส้นที่มีตัวแปรหลายตัว  
* สามารถเขียนแทนด้วยเมทริกซ์  
* **ตัวอย่าง:**

$2x + 3y = 5$  
$x - y = 1$

สามารถเขียนเป็น:

$
\begin{bmatrix}
2 & 3 \\
1 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
5 \\
1
\end{bmatrix}
$

**การประยุกต์ใช้ Linear Algebra ใน Machine Learning:**

* **การลดมิติข้อมูล (Dimensionality Reduction):**  เช่น Principal Component Analysis (PCA)  
    * PCA ใช้ Eigenvectors และ Eigenvalues ในการหาทิศทางที่มีความแปรปรวนสูงสุดในข้อมูล
    * ช่วยลดจำนวน Features ในขณะที่ยังคงรักษาข้อมูลสำคัญไว้ได้มากที่สุด

* **การวิเคราะห์รูปภาพ (Image Analysis):**  เช่น Eigenfaces  
    * Eigenfaces ใช้ Eigenvectors ในการแทนใบหน้า
    * ใช้ในการจดจำใบหน้า, การจัดกลุ่มใบหน้า

* **Natural Language Processing (NLP):**  เช่น Word Embeddings  
    * Word Embeddings ใช้ Vector ในการแทนความหมายของคำ
    * ใช้ในการแปลภาษา, การวิเคราะห์ความรู้สึก, การสร้าง Chatbot

## Bases

เราเขียน function เพื่อตรวจสอบว่าเป็นเวกเตอร์เป็น linear combination ของเวกเตอร์ที่กำหนดหรือไม่

```python
import numpy as np
import numpy.linalg

def is_linear_combination(vectors, target):
    matrix = np.array(vectors).T
    return np.linalg.matrix_rank(matrix) == np.linalg.matrix_rank(np.column_stack((matrix, target)))

vectors = [[1, 0, 1], [-1, 2, 1]]
target = [-5, 6, 1]

if is_linear_combination(vectors, target):
    print(f"{target} is a linear combination of {vectors}")
else:
    print(f"{target} is not a linear combination of {vectors}")

vectors = [[1, 0, 1], [-1, 2, 1]]
target = [-5, 6, 2]

if is_linear_combination(vectors, target):
    print(f"{target} is a linear combination of {vectors}")
else:
    print(f"{target} is not a linear combination of {vectors}")
```

เราเขียน function เพื่อตรวจสอบว่าเป็นเวกเตอร์ที่กำหนดให้ linearly independent กันหรือไม่

```python
import numpy as np
import numpy.linalg

def is_linearly_independent(vectors):
    matrix = np.array(vectors).T
    return np.linalg.matrix_rank(matrix) == len(vectors)

vectors = [[1, 4, 0], [10, 2, 1], [-5, 0 ,6]]

if is_linearly_independent(vectors):
    print(f"{vectors} are linearly independent")
else:
    print(f"{vectors} are linearly dependent")

vectors = [[4, 1, -2], [-3, 0, 1], [1, -2 , 1]]

if is_linearly_independent(vectors):
    print(f"{vectors} are linearly independent")
else:
    print(f"{vectors} are linearly dependent")
```

เราเขียน function เพื่อคำนวณเวกเตอร์ coordinate ของ basis ที่กำหนด

```python
import numpy as np
import numpy.linalg

def find_coordinate_vector(basis, vector):
    matrix = np.array(basis).T
    coefficients = np.linalg.solve(matrix, vector)
    return coefficients

basis = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]
vector = [1, 2, 3]

coordinate_vector = find_coordinate_vector(basis, vector)

print(f"The coordinate vector of {vector} relative to {basis} is {coordinate_vector}")

basis = [[1, 0, 0], [1, 1, 0], [1, 1, 1]]
vector = [1, 2, 3]

coordinate_vector = find_coordinate_vector(basis, vector)

print(f"The coordinate vector of {vector} relative to {basis} is {coordinate_vector}")
```

## Change of Basis

เราสร้างเมทริกซ์ transformation ด้วยการหาเวกเตอร์ coordinate ของ basis เก่าใน basis ใหม่

```python
import numpy as np
import numpy.linalg

def change_of_basis_matrix(new_basis, old_basis):
    new_basis = np.array(new_basis)
    old_basis = np.array(old_basis)

    if not (old_basis.shape[0] == old_basis.shape[1] == new_basis.shape[0] == new_basis.shape[1]):
        raise ValueError("Basis matrices must be square and have the same dimensions.")

    try:
        inv_new_basis = np.linalg.inv(new_basis)
    except np.linalg.LinAlgError:
        raise ValueError("The inverse of the new basis matrix does not exist.")

    transformation_matrix = np.dot(inv_new_basis, old_basis)

    return transformation_matrix

old_basis_matrix = np.array([[1, 0, 0],
                             [0, 1, 0],
                             [0, 0, 1]])

new_basis_matrix = np.array([[1, 1, 1],
                             [0, 1, 1],
                             [0, 0, 1]])

transformation_matrix = change_of_basis_matrix(new_basis_matrix, old_basis_matrix)
print("Transformation matrix:")
print(transformation_matrix)
```

เอาเมทริกซ์ transformation ไปคูณกับเมทริกซ์ที่ต้องการเปลี่ยน basis

```python
import numpy as np

old_coordinate_vector = [-1, -1, 3]

# หา coordinate ของเวกเตอร์ใน basis ใหม่
new_coordinate_vector = transformation_matrix @ old_coordinate_vector

print("Matrix A:")
print(old_coordinate_vector)

print("\nMatrix B:")
print(new_coordinate_vector)
```

## Principal Component Analysis

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
data = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])

x = data.iloc[:,0:4]
target = data.iloc[:,4]

display(data)
```

eigenvalue และ eigenvector สำหรับเมทริกซ์สมมาตรหาได้ด้วยฟังก์ชัน np.linalg.eigh()

```python
import numpy as np

def PCA(X , num_components):
    X_meaned = X - np.mean(X , axis = 0)
    cov_mat = np.cov(X_meaned , rowvar = False)

    # หา eigenvalues และ eigenvectors ของเมทริกซ์ covariance
    eigenvalues, eigenvectors = np.linalg.eigh(cov_mat)

    sorted_index = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalue = eigenvalues[sorted_index]
    sorted_eigenvectors = eigenvectors[:,sorted_index]

    eigenvector_subset = sorted_eigenvectors[:,0:num_components]

    X_reduced = np.dot(eigenvector_subset.T, X_meaned.T).T

    return X_reduced
```

```python
x_reduced = PCA(x,2)

principal_df = pd.DataFrame(x_reduced , columns = ['PC1','PC2'])
principal_df = pd.concat([principal_df , pd.DataFrame(target)] , axis = 1)

print(principal_df)

import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize = (6,6))
sb.scatterplot(data = principal_df, x = 'PC1', y = 'PC2', hue = 'target', s = 60 )
```

## Singular Value Decomposition

```python
import requests
import numpy as np
from PIL import Image
from io import BytesIO

url = 'https://cms-assets.skooldio.com/blog/wp-content/uploads/2023/05/16192037/346836943_936895414241613_4860047313685873918_n.jpg'
response = requests.get(url)
image = Image.open(BytesIO(response.content)).convert('L')

image_array = np.array(image)

U, s, VT = np.linalg.svd(image_array, full_matrices=False)

k=100

# หาเมทริกซ์รูปภาพด้วยการคูณบางส่วนของเมทริกซ์ U s และ VT เข้าด้วยกัน
reconstructed_image = U[:, :k] @ np.diag(s[:k]) @ VT[:k,:]

reconstructed_image = Image.fromarray(reconstructed_image.astype('uint8'))

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(reconstructed_image, cmap='gray')
axes[1].set_title(f'Reconstructed Image')
axes[1].axis('off')
plt.show()
```

# Differential Calculus
**แคลคูลัสเชิงอนุพันธ์ (Differential Calculus)** คือ สาขาของคณิตศาสตร์ที่ศึกษาเกี่ยวกับอัตราการเปลี่ยนแปลง  มีบทบาทสำคัญในการหาค่าต่ำสุดของฟังก์ชัน ซึ่งเป็นพื้นฐานของ Optimization ใน Machine Learning

**อนุพันธ์ (Derivative):**  อัตราการเปลี่ยนแปลง ณ จุดใดจุดหนึ่ง  
* **ความชันของเส้นสัมผัส (Tangent Line):**  อนุพันธ์ของฟังก์ชัน ณ จุด $x$ คือ ความชันของเส้นสัมผัสกราฟ ณ จุด $(x,  f(x))$
* **สัญลักษณ์:**  $f'(x)$,  $\frac{df}{dx}$,  $\frac{dy}{dx}$

**การหาอนุพันธ์:**

* **อนุพันธ์ของฟังก์ชันพหุนาม:**  
    * $f(x) = x^n$ ,  $f'(x) = nx^{n-1}$  
    * **ตัวอย่าง:**  $f(x) = x^3$ ,  $f'(x) = 3x^2$

* **กฎลูกโซ่ (Chain Rule):**  ใช้หาอนุพันธ์ของฟังก์ชันประกอบ  
    * $f(g(x))' = f'(g(x))g'(x)$  
    * **ตัวอย่าง:**  $f(x) = sin(x^2)$ ,  $f'(x) = cos(x^2) \cdot 2x$

* **กฎผลคูณ (Product Rule):**  ใช้หาอนุพันธ์ของผลคูณของฟังก์ชัน  
    * $(u(x)v(x))' = u'(x)v(x) + u(x)v'(x)$  
    * **ตัวอย่าง:**  $f(x) = x^2 e^x$ ,  $f'(x) = 2xe^x + x^2e^x$

* **กฎผลหาร (Quotient Rule):**  ใช้หาอนุพันธ์ของผลหารของฟังก์ชัน  
    * $( \frac{u(x)}{v(x)})' = \frac{u'(x)v(x) - u(x)v'(x)}{v(x)^2}$
    * **ตัวอย่าง:**
        \begin{align*}
        f(x) &= \frac{x}{x^2 + 1},  \\
        f'(x) &= \frac{(1)(x^2 + 1) - (x)(2x)}{(x^2 + 1)^2} \\
        &= \frac{1 - x^2}{(x^2 + 1)^2}
        \end{align*}

**การประยุกต์ใช้ Differential Calculus ใน Machine Learning:**

* **Gradient Descent:**  Algorithm สำหรับหาค่าต่ำสุดของฟังก์ชัน  
    * ใช้ Gradient (เวกเตอร์ของอนุพันธ์ย่อย) ในการหาทิศทางที่ลดลงเร็วที่สุด
    * ใช้ในการหาค่า Parameter ที่เหมาะสมที่สุดของ Model

* **Backpropagation:**  Algorithm สำหรับปรับปรุง Neural Network  
    * ใช้ Chain Rule ในการคำนวณ Gradient ของ Loss Function
    * ใช้ในการปรับ Weight และ Bias ของ Neural Network

* **Optimization:**  การหาค่า Parameter ที่เหมาะสมที่สุด  
    * ใช้ Differential Calculus ในการหาจุดต่ำสุดของ Loss Function
    * มีบทบาทสำคัญในการสร้าง Model ที่มีประสิทธิภาพ


## Interpolation

เราทำ Lagrange interpolation ด้วยฟังก์ชัน lagrange ใน scipy.interpolate



```python
import numpy as np
from scipy.interpolate import lagrange
from numpy.polynomial.polynomial import Polynomial
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 12)
y = np.exp(np.square(x))

# หา Lagrange interpolation
poly = lagrange(x,y)

x_new = np.linspace(-1, 1, 300)
plt.scatter(x, y, label='data', color='black')
plt.plot(x_new, Polynomial(poly.coef[::-1])(x_new), label='interpolated function', color='red')
plt.plot(x_new, np.exp(np.square(x_new)), label=r"$real function$", linestyle='-.', color='blue')
plt.legend()
plt.show()
```

## Optimization Problem

```python
import time
import numpy as np

def objective_function(x):
    return x ** 4 + 4 * x

def diff(x):
    return 4 * x ** 3 + 4

# Grid search method
def grid_search():
    search_space = np.linspace(-10, 10, 1000)
    best_value = float('inf')
    best_x = None
    for x in search_space:
        value = objective_function(x)
        if value < best_value:
            best_value = value
            best_x = x
    return best_x, best_value

# Random search method
def random_search():
    np.random.seed(42)
    num_iterations = 1000
    best_value = float('inf')
    best_x = None
    for _ in range(num_iterations):
        x = np.random.uniform(-10, 10)
        value = objective_function(x)
        if value < best_value:
            best_value = value
            best_x = x
    return best_x, best_value

# Golden ratio search method
def golden_ratio_search():
    epsilon = 1e-6
    a, b = -10, 10
    phi = (1 + np.sqrt(5)) / 2
    x1 = b - (b - a) / phi
    x2 = a + (b - a) / phi
    while abs(b - a) > epsilon:
        if objective_function(x1) < objective_function(x2):
            b = x2
            x2 = x1
            x1 = b - (b - a) / phi
        else:
            a = x1
            x1 = x2
            x2 = a + (b - a) / phi
    return (a + b) / 2, objective_function((a + b) / 2)

# Descent method
def descent(learning_rate=0.1, epsilon=1e-6):
    x = 0  # Initial guess
    prev_value = objective_function(x)

    while True:
        x = x - learning_rate * diff(x)
        current_value = objective_function(x)
        if abs(current_value - prev_value) < epsilon:
            break
        prev_value = current_value

    return x, current_value

methods = {
    'Grid Search': grid_search,
    'Random Search': random_search,
    'Golden Ratio Search': golden_ratio_search,
    'Descent': descent
}

print("Optimization Method \t Time (s) \t Best X \t Best Value")
print("=" * 60)

for method_name, method_func in methods.items():
    start_time = time.time()
    best_x, best_value = method_func()
    end_time = time.time()
    execution_time = end_time - start_time
    print(f"{method_name:20} \t {execution_time:.6f} \t {best_x:.6f} \t {best_value:.6f}")
```

##Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt

def mean_squared_error(y_true, y_predicted):
  cost = np.sum((y_true-y_predicted)**2) / len(y_true)
  return cost

def gradient_descent(x, y, iterations = 1000, learning_rate = 0.0001, stopping_threshold=10e-3):

  current_slope = 0.1
  current_intercept = 0.01
  iterations = iterations
  learning_rate = learning_rate
  n = float(len(x))

  costs = []
  slopes = []
  previous_cost = None

  for i in range(iterations):
    y_predicted = (current_slope * x) + current_intercept
    current_cost = mean_squared_error(y, y_predicted)

    if previous_cost and abs(previous_cost-current_cost)<=stopping_threshold:
      break

    previous_cost = current_cost
    costs.append(current_cost)
    slopes.append(current_slope)

    slope_derivative = -(2/n) * sum(x * (y-y_predicted))
    intercept_derivative = -(2/n) * sum(y-y_predicted)

    current_slope = current_slope - (learning_rate * slope_derivative)
    current_intercept = current_intercept - (learning_rate * intercept_derivative)

    print(f"Iteration {i+1}: Cost {current_cost}, Slope {current_slope}, Intercept {current_intercept}")

  plt.figure(figsize = (6,4))
  plt.plot(range(i), costs, color='black')
  plt.scatter(range(i), costs, marker='o', color='red')
  plt.ylabel("Cost")
  plt.xlabel("Iteration")
  plt.show()

  plt.figure(figsize = (6,4))
  plt.plot(range(i), slopes, color='black')
  plt.scatter(range(i), slopes, marker='o', color='red')
  plt.ylabel("Slope")
  plt.xlabel("Iteration")
  plt.show()

  return current_slope, current_intercept
```

```python
X = np.array([32.50234527, 53.42680403, 61.53035803, 47.47563963, 59.81320787,
  55.14218841, 52.21179669, 39.29956669, 48.10504169, 52.55001444,
  45.41973014, 54.35163488, 44.1640495 , 58.16847072, 56.72720806,
  48.95588857, 44.68719623, 60.29732685, 45.61864377, 38.81681754])
Y = np.array([31.70700585, 68.77759598, 62.5623823 , 71.54663223, 87.23092513,
  78.21151827, 79.64197305, 59.17148932, 75.3312423 , 71.30087989,
  55.16567715, 82.47884676, 62.00892325, 75.39287043, 81.43619216,
  60.72360244, 82.89250373, 97.37989686, 48.84715332, 56.87721319])

estimated_slope, estimated_intercept = gradient_descent(X, Y, iterations=2000)
print(f"Best-Fit Line: y = {estimated_slope:.3f}x + ({estimated_intercept:.3f})")

Y_pred = estimated_slope * X + estimated_intercept

plt.figure(figsize = (6,4))
plt.scatter(X, Y, marker='o', color='black')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red',markerfacecolor='red', markersize=10)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
